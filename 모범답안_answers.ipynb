## 과제 C (심화): 필요 데이터 & AI 방법론

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic'
import matplotlib.pyplot as plt

필요_데이터 = ['온도 센서', '습도 센서', '분사 압력', '도료 점도', '불량 여부(레이블)', '작업자 ID']
ai_방법 = {
    '레이블 있음 (지도학습)': 'Random Forest / XGBoost 분류',
    '레이블 없음 (비지도)':   'Isolation Forest 이상탐지',
    '시계열 패턴':           'LSTM 이상 예측',
}

print("필요 데이터:")
for d in 필요_데이터:
    print(f"  - {d}")
print()
print("AI 방법론 선택:")
for cond, method in ai_방법.items():
    print(f"  [{cond}] → {method}")
print()
print("→ 도장 공정 초기: 레이블(불량 판정) 수집 비용이 높으므로")
print("  Isolation Forest(비지도)로 시작 후 레이블 축적 시 지도학습 전환 권장")
```
